We can use the following command for detection,

yolo task=detect mode=predict model="path/to/best.pt" source="Image File Path" device="CPU / 0,1,2,etc"

Here,

model = Location (path) of the trained model’s {best.pt}

source = it can be any image, video file, image/video on the net, etc.

device = CPU or GPU you want to run the detections on

Example:

!yolo task=detect mode=predict model="/content/runs/detect/train/weights/best.pt" source="/content/PCB-Dataset-Defect-1/test/images/10_short_05_jpg.rf.fa4626030817967e6cf0e58874325530.jpg" device=0
